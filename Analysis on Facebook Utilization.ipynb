# **6. Data Pre-Processing**

- This section is emphasised on performing data manipulation over unstructured data for further processing and analysis.

```python
clean_data = data.copy()
```

##### Handling missing data

```python
clean_data.isna().sum()
```

```python
clean_data['gender'].unique()   # find the nan value
```

```python
# so we need to put 'Unkonwn' or other related/identified value of gender instead NAN value.
# used fillna with Unknown as of now.
clean_data['gender'] = clean_data['gender'].fillna("UnKnown")
```

```python
clean_data['gender'].unique()  # checked fillout "Unknown" instad nan.
```

```python
clean_data[clean_data['tenure'].isna()]
```

```python
# the use has like dislikes and friend and friendship etc so we deside to go with the mean value to fillna

clean_data['tenure'] = clean_data['tenure'].fillna(clean_data['tenure'].mean())
```

# **8. Exploratory Data Analysis**
---

- This section is emphasised on asking the right questions and perform analysis using the data.

- Note that there is no limit how deep you can go, but make sure not to get distracted from right track.

```python
clean_data.columns
```

```python
clean_data['gender'].unique()
```

```python
# Initialize figure of size 15 X 7
fig = plt.figure(figsize=(15,4))

labels = ['Male', 'Female', 'Unknown']
values = clean_data['gender'].value_counts()
colors = ['lightblue', 'lightgreen', 'lightcoral']
explode = (0.1, 0.1, 0.1)  # To explode the first slice (HomeTeam)

plt.pie(values, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=45)
plt.axis('equal')  # Equal aspect ratio ensures that the pie chart is circular.

plt.title('Gender who used Facebook\n')
plt.savefig("pieoutput")
plt.show()
```

```python
clean_data['age'].unique()
```

```python
# Age Groups Using Facebook

TotalAge  = clean_data.groupby('age').count()

sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.lineplot(data=TotalAge, x='age', y='userid', marker='o')
plt.xticks(rotation=45)
plt.title('Age Groups Using Facebook')
plt.xlabel('Age')
plt.ylabel('Facebook uses Count')
plt.savefig("facebookusescount")
plt.show()
```

```python
clean_data.columns
```

```python
figure, ax = plt.subplots(figsize=(15, 7))
agegrp = clean_data.groupby('age').agg({'www_likes':'mean','mobile_likes':'mean'})
sns.lineplot(x='age', y='www_likes', data=agegrp, ax=ax, label='Facebook Average Website Users')
sns.lineplot(x='age', y='mobile_likes', data=agegrp, ax=ax, label='Facebook Average Mobile Users')
plt.xticks(rotation=45)
plt.xlabel('Age')
plt.ylabel('Average Users')
plt.title('Website Vs Mobile uses on Facebook')
plt.legend()
plt.tight_layout()
plt.savefig("fbwebmobile")
plt.show()
```

```python
fbavgfrnd = clean_data.groupby('age').agg({'friend_count':'mean'})
    
figure, ax = plt.subplots(figsize=(15, 7))
fbavgfrnd.plot(kind='bar', ax=ax, color='darkorange', title='Agerage Facebook Friends')
plt.tight_layout()
plt.savefig("avgfbfriends")
plt.show()
```

```python

wwlikerec = clean_data.groupby('age')['www_likes_received'].mean()
moblikerec = clean_data.groupby('age')['mobile_likes_received'].mean()

# Combine both average goals into a single DataFrame
totallikerec = pd.DataFrame({
    'wwlike': wwlikerec.index,
    'wwlikeavg': wwlikerec.values,
    'moblike': moblikerec.index,
    'moblikeavg': moblikerec.values
})

# team_avg_goals['Overalllikerecv'] = (team_avg_goals['wwlikeavg'] + team_avg_goals['moblikeavg']) / 2
# team_avg_goals_sorted = team_avg_goals.sort_values(by='Overalllikerecv', ascending=False)

# Create the bar plot with sorted data
plt.figure(figsize=(15, 7))
plt.bar(totallikerec['wwlike'], totallikerec['wwlikeavg'], color='skyblue', label='Website Users like Recieved')
plt.bar(totallikerec['moblike'], totallikerec['moblikeavg'], color='darkorange', label='Mobile Users like Recieved')
plt.xlabel('Age Group Users')
plt.ylabel('Average Like Recieved')
plt.title('Average Like per Age Group')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.savefig("avglikerec")
plt.show()
```
